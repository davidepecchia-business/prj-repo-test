```sas
title 'Test bar chart';
proc sgpanel data=public.hmeq(where=(JOB in ('Sales' 'Mgr'))) noautolegend;
  panelby JOB / novarname columns=2 onepanel;
  vbar BAD / response=LOAN stat=mean;
  rowaxis grid;
  run;
```
